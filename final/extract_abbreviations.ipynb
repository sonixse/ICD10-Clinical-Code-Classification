```python
import re
import pandas as pd
from collections import Counter
```

```python
test_df     = pd.read_csv('data/raw/test_leaderboard_data.csv')
training_df = pd.read_csv('data/raw/training_codification_data.csv')

# literals = pd.concat([test_df['Literal'], training_df['Literal']]).dropna()
literals = training_df['Literal'].dropna()

ABB_RE = re.compile(r'\b[A-Z]{2,5}(?:\d)?\b')

found = Counter()
for lit in literals:
    for match in ABB_RE.findall(str(lit)):
        found[match] += 1

ICD_RE = re.compile(r'^[A-Z]\d')
abbreviations = {k: v for k, v in found.items() if not ICD_RE.match(k)}

print(f"Found {len(abbreviations)} abbreviations:\n")
for abbr, count in sorted(abbreviations.items(), key=lambda x: -x[1]):
    print(f"  {abbr:<10} {count:>4}x")
```
