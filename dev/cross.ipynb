```python
import pandas as pd

icd_d_p_pairs_df = pd.read_csv("data/raw/icd_d_p_pairs.csv")
trainig_condification_data_df = pd.read_csv("data/raw/training_codification_data.csv")
```

```python
# Cross-reference rows present in both dataframes using the "Code" column
cross_referenced_df = pd.merge(
    icd_d_p_pairs_df,
    trainig_condification_data_df,
    on="Code",
    how="inner",
    suffixes=("_icd", "_training")
)

print(f"Matched rows: {len(cross_referenced_df)}")
cross_referenced_df.head()
```

```python
#save the cross-referenced dataframe to a new CSV file
cross_referenced_df.to_csv("data/utils/cross_referenced_data.csv", index=False)
```

```python
# Cross-reference rows present in both dataframes using the "Code" column now for processed data
p_icd = pd.read_csv('data/preprocessed/icd_preprocessed_descriptions.csv')
p_train = pd.read_csv("data/preprocessed/preprocessed_data.csv")

preprocessed_cross = pd.merge(
    p_icd,
    p_train,
    on="Code",
    how="inner",
    suffixes=("_Description", "_Literal")
).loc[:, ["Code", "D_P", "processed_Description", "processed_Literal"]]

print(f"Matched rows: {len(preprocessed_cross)}")
preprocessed_cross.head()
```

```python
preprocessed_cross.to_csv("data/utils/preprocessed_cross.csv", index=False)
```
